##### a.	Assuming linear relationships, plot the band gap as a function of each of the other properties (melting temperature, covalent radius, bond energy, and first ionization energy) and predict the band gap of diamond.!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Carbon-group-properties.csv", header=0, index_col=0)

y = data.iloc[-1, 2:].values
melt_tem = data.iloc[0, 2:].values
radius = data.iloc[1, 2:].values
bond_energy = data.iloc[2, 2:].values
ioni_energy = data.iloc[3, 2:].values
```

```python
phi_m = np.zeros((y.shape[0], 2))

phi_m[:, 0] = np.ones(y.shape[0])
phi_m[:, 1] = melt_tem

theta_ls_melt_tem = np.linalg.lstsq(phi_m.astype(float), y.astype(float), rcond=None)[0]

# print(theta_ls_melt_tem)

x_fit = np.linspace(min(melt_tem), max(melt_tem), 100)
y_fit = theta_ls_melt_tem[0]+theta_ls_melt_tem[1]*x_fit

plt.figure(figsize=[5, 5])
plt.scatter(melt_tem, y)
plt.plot(x_fit, y_fit, color="red")
# plt.xscale('log')
# plt.yscale('log')
plt.title("Band Gap & melting temperature")
plt.xlabel("Melting temperature")
plt.ylabel("Band Gap")
plt.show()
```

```python
phi_r = np.zeros((y.shape[0], 2))

phi_r[:, 0] = np.ones(y.shape[0])
phi_r[:, 1] = radius

theta_ls_radius= np.linalg.lstsq(phi_r.astype(float), y.astype(float), rcond=None)[0]

# print(theta_ls_melt_tem)

x_fit_radius = np.linspace(min(radius), max(radius), 100)
y_fit_radius = theta_ls_radius[0]+theta_ls_radius[1]*x_fit_radius

plt.figure(figsize=[5, 5])
plt.scatter(radius, y)
plt.plot(x_fit_radius, y_fit_radius, color="red")
# plt.xscale('log')
# plt.yscale('log')
plt.title("Band Gap & radius")
plt.xlabel("radius")
plt.ylabel("Band Gap")
plt.savefig("Band Gap & radius", bbox_inches="tight")
```

```python
phi_b = np.zeros((y.shape[0], 2))

phi_b[:, 0] = np.ones(y.shape[0])
phi_b[:, 1] = bond_energy

theta_ls_bond = np.linalg.lstsq(phi_b.astype(float), y.astype(float), rcond=None)[0]

# print(theta_ls_bond)

x_fit_bond = np.linspace(min(bond_energy), max(bond_energy), 100)
y_fit_bond = theta_ls_bond[0]+theta_ls_bond[1]*x_fit_bond

plt.figure(figsize=[5, 5])
plt.scatter(bond_energy, y)
plt.plot(x_fit_bond, y_fit_bond, color="red")
# plt.xscale('log')
# plt.yscale('log')
plt.title("Band Gap & bond_energy")
plt.xlabel("bond_energy")
plt.ylabel("Band Gap")
plt.savefig("Band Gap & bond_energy", bbox_inches="tight")
```

```python
phi_i = np.zeros((y.shape[0], 2))

phi_i[:, 0] = np.ones(y.shape[0])
phi_i[:, 1] = ioni_energy

theta_ls_ioni= np.linalg.lstsq(phi_i.astype(float), y.astype(float), rcond=None)[0]

# print(theta_ls_bond)

x_fit_ioni = np.linspace(min(ioni_energy), max(ioni_energy), 100)
y_fit_ioni = theta_ls_ioni[0]+theta_ls_ioni[1]*x_fit_ioni

plt.figure(figsize=[5, 5])
plt.scatter(ioni_energy, y)
plt.plot(x_fit_ioni, y_fit_ioni, color="red")
# plt.xscale('log')
# plt.yscale('log')
plt.title("Band Gap & ionic_energy")
plt.xlabel("ionic_energy")
plt.ylabel("Band Gap")
plt.savefig("Band Gap & ionic_energy", bbox_inches="tight")
```
